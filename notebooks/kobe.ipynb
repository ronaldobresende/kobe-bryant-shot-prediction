```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os datasets
dev_data = pd.read_parquet("c:/projects/python/kobe-bryant-shot-prediction/data/01_raw/dataset_kobe_dev.parquet")
prod_data = pd.read_parquet("c:/projects/python/kobe-bryant-shot-prediction/data/01_raw/dataset_kobe_prod.parquet")

# Comparar distribuições de uma variável (exemplo: 'shot_distance')
sns.kdeplot(dev_data['shot_distance'], label='Dev', shade=True)
sns.kdeplot(prod_data['shot_distance'], label='Prod', shade=True)
plt.legend()
plt.title("Distribuição da variável 'shot_distance'")
plt.show()
```

```python
print("Valores ausentes no dataset de desenvolvimento:")
print(dev_data.isnull().sum())

print("\nValores ausentes no dataset de produção:")
print(prod_data.isnull().sum())
```

```python
print("Proporção da variável alvo no dataset de desenvolvimento:")
print(dev_data['shot_made_flag'].value_counts(normalize=True))

print("\nProporção da variável alvo no dataset de produção:")
print(prod_data['shot_made_flag'].value_counts(normalize=True))
```

```python
# Proporção de valores nulos na variável 'shot_made_flag' no dataset de produção
null_proportion = prod_data['shot_made_flag'].isnull().mean()
print(f"Proporção de valores nulos em 'shot_made_flag' no dataset de produção: {null_proportion:.2%}")
```
